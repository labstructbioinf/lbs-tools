```python
# taxids of genomes that are sequenced
genome_taxids = genomes.taxid.tolist()

df = tax.nodes.drop(tax.nodes.index[(tax.nodes['rank'] == 'species') & (~tax.nodes['taxid'].isin(genome_taxids))])

len(df[df['rank']=='species'])
```

```python
def getchildren(taxid):
    """
    get all leafs (organisms) that are descendants of the given taxid
    """

    children = df[df.parent_taxid == taxid].taxid.tolist()

    if len(children)==0:
        if df[df.taxid==taxid]['rank'].iloc[0] == 'species':
            print(taxid)
    else:
        for child in children:
            getchildren(child)
```
