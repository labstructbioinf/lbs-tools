## Example of usage

#### Selecting some enzyme of interest:

```python
df = pdb.ecod[pdb.ecod['t_name'] == 'Thymidylate synthase/dCMP hydroxymethylase']
```

```python
df.head()
```

#### Now select only those containing single ligand type

```python
df = df[df['ligand'] == 'UMP']
```

```python
df.head()
```

```python
len(df), len(df['pdb_id'].unique())
```

#### Merge with pdb.chains DF

```python
df = pd.merge(pdb.chains.loc[df.pdb_id], df,  left_index=True, right_on='pdb_id')
```

#### Filter and sort by resolution

```python
df = df[df['resolution'] <= 2.5].sort_values(by='resolution')
```

#### Get unique PDB structures (not chains)

```python
df['pdb'] = df['pdb_id'].apply(lambda x: x.split('_')[0])
```

```python
ids = df.groupby(by='pdb').idxmin().index.tolist()
pdb.structures.loc[ids]
```

## Another example

```python
import pandas as pd
from lbs.localpdb import PDB
pdb = PDB(db_path='/home/db/pdb', common_filters=True, plugins=['ECOD'])
```

```python
pdb.load_ecod_data('/home/users/jludwiczak/ecod.latest.domains.txt')
```

#### Get only Rossman-like

```python
pdb.ecod = pdb.ecod[pdb.ecod['x_name'] == 'Rossmann-like']
```

```python
len(pdb.ecod)
```

```python
pdb.ecod['has_SAM'] = pdb.ecod['ligand'].apply(lambda x: 'SAM' in str(x))
pdb.ecod = pdb.ecod[pdb.ecod['has_SAM'] == True]
del pdb.ecod['has_SAM']
len(pdb.ecod)
```

```python
pdb.ecod.head()
```

#### Collapse to single pdb_chain

```python
len(pdb.ecod)
```

```python
def group_func(x):
    data = {}
    for ecod_domain_id, _id, pdb_range, seqid_range in zip(x.index, x['ecod_number'], x['pdb_range'], x['seqid_range']):
        data[ecod_domain_id] = [_id, pdb_range, seqid_range]
    return data
```

```python
pdb.ecod.groupby('pdb_id').apply(group_func).head()
```

```python
len(pdb.ecod.groupby('pdb_id').apply(group_func))
```

#### Example of collapsed pdb_chain with 2 domains

```python
for key, value in pdb.ecod.groupby('pdb_id').apply(group_func).iteritems():
    if len(value) > 1:
        print(key, value)
        break
```
